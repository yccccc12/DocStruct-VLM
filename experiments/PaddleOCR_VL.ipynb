```python
from paddleocr import PaddleOCRVL

pipeline = PaddleOCRVL()
```

```python
# 85s
output = pipeline.predict("/content/Test1.png")

for res in output:
    res.print()
    res.save_to_img(save_path="output")
    res.save_to_json(save_path="output")
    res.save_to_markdown(save_path="output")
```

```python
input_file = "/content/UECM3993_MAY2025-1-3.pdf"
output_file = pipeline.predict(input=input_file)
```

```python
pages_res = list(output_file)
pages_res
```

```python
output_2 = pipeline.restructure_pages(pages_res)
# output = pipeline.restructure_pages(pages_res, merge_tables=True) # Merge tables across pages
# output = pipeline.restructure_pages(pages_res, merge_tables=True, relevel_titles=True) # Merge tables across pages and reconstruct multi-level titles
# output = pipeline.restructure_pages(pages_res, merge_tables=True, relevel_titles=True, concatenate_pages=True) # Merge tables across pages, reconstruct multi-level titles, and merge multiple pages

for res in output_2:
    res.print() ## Print the structured prediction output
    res.save_to_img(save_path="output_2") ## Save the current image's structured result in PNG format
    res.save_to_json(save_path="output_2") ## Save the current image's structured result in JSON format
    res.save_to_markdown(save_path="output_2") ## Save the current image's result in Markdown format
```

```python
from IPython.display import display
for res in output_2:
    layout_img = res.img["layout_det_res"]  # PIL Image with bounding boxes
    display(layout_img)
```

```python
# output_2 = pipeline.restructure_pages(pages_res)
# output_2 = pipeline.restructure_pages(pages_res, merge_tables=True) # Merge tables across pages
# output_2 = pipeline.restructure_pages(pages_res, merge_tables=True, relevel_titles=True) # Merge tables across pages and reconstruct multi-level titles
output_2 = pipeline.restructure_pages(pages_res, merge_tables=True, relevel_titles=True, concatenate_pages=True) # Merge tables across pages, reconstruct multi-level titles, and merge multiple pages

for res in output_2:
  res.save_to_img("output_restructured")
  res.save_to_json(save_path="output_restructured")
  res.save_to_markdown("output_restructured")
```
