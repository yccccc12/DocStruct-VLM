```python
from transformers import AutoModel, AutoTokenizer
import torch, os
os.environ["CUDA_VISIBLE_DEVICES"] = "0"
model_name = "deepseek-ai/DeepSeek-OCR-2"
tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
model = AutoModel.from_pretrained(
    model_name,
    trust_remote_code=True,
    low_cpu_mem_usage=True,
    torch_dtype=torch.bfloat16,
    device_map="cuda",
)

model = model.eval()
```

```python
import os
import glob
import subprocess

# --- CONFIGURATION ---
DATASET_TYPE = "handwritten_en"   # Change to "handwritten_en" for English dataset
MODEL_NAME = "deepseekOCR2"       # Model identifier for output organization

# --- DYNAMIC PATH SETTINGS ---
INPUT_FOLDER = os.path.join("data", "raw", DATASET_TYPE, "images")
BASE_OUTPUT_DIR = os.path.join("outputs", DATASET_TYPE, MODEL_NAME)


# --- PROMPT ---
PROMPT = "<image>\n<|grounding|>Convert the image to markdown. "

def run_deepseekocr(file_path, output_dir):
    """
    Runs deepseekocr on a file.

    Args:
        file_path (str): Path to image or PDF.
        output_dir (str): Directory to save output.
    """

    file_path = os.path.abspath(file_path)
    output_dir = os.path.abspath(output_dir)

    file_name = os.path.splitext(os.path.basename(file_path))[0]
    output_path = os.path.join(output_dir, f"{file_name}")

    try:
        res = model.infer(
            tokenizer,
            prompt=PROMPT,
            image_file=file_path,
            output_path=output_path,
            base_size=1024,
            image_size=768,
            crop_mode=True,
            save_results=True
        )

        # Optional: print short preview
        if isinstance(res, str):
            print(res[:100])

        print(f"Saved → {output_path}")

    except Exception as e:
        print(f"❌ Error processing {file_path}: {e}")

def run_batch_ocr():
    # Gather image and PDF files
    file_extensions = ("*.png", "*.jpg", "*.jpeg", "*.pdf")
    file_list = []

    for ext in file_extensions:
        file_list.extend(glob.glob(os.path.join(INPUT_FOLDER, ext)))

    if not file_list:
        print(f"No files found in: {os.path.abspath(INPUT_FOLDER)}")
        return

    # Process files
    for i, file_path in enumerate(file_list):
        file_name = os.path.basename(file_path)

        print(f"\n[{i+1}/{len(file_list)}] Processing: {file_name}")

        run_deepseekocr(file_path, BASE_OUTPUT_DIR)

    print("\n" + "="*40)
    print(f"Batch processing complete!")
    print("="*40)

if __name__ == "__main__":
    run_batch_ocr()
```

```python
import fitz, io
from PIL import Image

def pdf_to_images(pdf_path, dpi=144):
    doc = fitz.open(pdf_path)
    zoom = dpi / 72.0
    matrix = fitz.Matrix(zoom, zoom)
    images = []
    for page in doc:
        pix = page.get_pixmap(matrix=matrix, alpha=False)
        img = Image.open(io.BytesIO(pix.tobytes("png")))
        if img.mode in ("RGBA", "LA"):
            bg = Image.new("RGB", img.size, (255, 255, 255))
            bg.paste(img, mask=img.split()[-1] if img.mode == "RGBA" else None)
            img = bg
        images.append(img)
    doc.close()
    return images
```
